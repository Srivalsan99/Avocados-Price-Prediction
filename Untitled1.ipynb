```python
def num(x):
  plt.figure(figsize=(12,5))
  ax1=plt.subplot(1,3,1)
  sns.distplot(x,ax=ax1)
  ax2=plt.subplot(1,3,2)
  sns.boxplot(x,ax=ax2)
  ax3=plt.subplot(1,3,3)
  sns.violinplot(x,ax=ax3)
  plt.show()

def cat(x):
  plt.figure(figsize=(20,10))
  ax1=plt.subplot(1,2,1)
  sns.countplot(x,ax=ax1)
  plt.xticks(rotation=90)
  ax2=plt.subplot(1,2,2)
  x.value_counts().plot(kind='pie',autopct='%1.1f%%',ax=ax2)
  plt.xticks(rotation=90)
  plt.show()
```

```python
# Building a model

from sklearn.linear_model import LinearRegression,SGDRegressor
```

```python
df[['2015','2016','2017','2018']]=pd.get_dummies(df['year'])
```

```python
# Region is dropped
df.drop(columns='region',inplace=True)
```

```python
df[['conventional','organic']]=pd.get_dummies(df['type'])
```

```python
df.drop(columns=['type','Date'],inplace=True)
```

```python
x=df.drop(columns='AveragePrice')
y=df['AveragePrice']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,shuffle=True,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train_scaled=pd.DataFrame(ss.fit_transform(x_train),columns=x_train.columns)
x_test_scaled=pd.DataFrame(ss.transform(x_test),columns=x_test.columns)
```

**Linear Models**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train_scaled,y_train)
y_train_pred_lr=lr.predict(x_train_scaled)
y_test_pred_lr=lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error
print('R2  train:', lr.score(x_train_scaled,y_train))
print('R2 test:', lr.score(x_test_scaled,y_test))
```

```python
rmse_train =  np.sqrt(mean_squared_error(y_train,y_train_pred_lr))
rmse_test = np.sqrt(mean_squared_error(y_test,y_test_pred_lr))
print('RMSE train:',rmse_train)
print('RMSE test:',rmse_test)
```

```python
from sklearn.linear_model import SGDRegressor
sgd=SGDRegressor()
sgd.fit(x_train_scaled,y_train)
y_train_pred_sgd=sgd.predict(x_train_scaled)
y_test_pred_sgd=sgd.predict(x_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error
print('R2  train:', sgd.score(x_train_scaled,y_train))
print('R2 test:', sgd.score(x_test_scaled,y_test))
```

```python
rmse_train =  np.sqrt(mean_squared_error(y_train,y_train_pred_sgd))
rmse_test = np.sqrt(mean_squared_error(y_test,y_test_pred_sgd))
print('RMSE train:',rmse_train)
print('RMSE test:',rmse_test)
```
